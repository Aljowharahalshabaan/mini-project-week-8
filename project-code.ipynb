```python
"""
Day 40 Activity: Matrix Operations Exercise
Tasks:
1) Build a simple linear transformation Y = X @ W + b
2) Avoid loops; use vectorized operations
3) Print shapes and verify output
"""

import numpy as np

# Provided inputs (edit if you want different values)
X = np.array([
    [1.0, 0.5],
    [2.0, -1.0],
    [0.0, 3.0],
])
W = np.array([
    [0.2, -0.1, 0.5],
    [0.7, 0.3, -0.2],
])
b = np.array([0.1, 0.0, -0.3])

# TODO: Compute Y = X @ W + b
#downnnnn👇👇
# TODO: Print shapes
print("original X: ",X.shape)
print("original W: ",W.shape)
print("original b: ",b.shape)
```

```python
means = X.mean(axis=0, keepdims=True)
stds = X.std(axis=0, keepdims=True) 
X_standardized= (X - means) / stds
print("X_standardized: ",X_standardized.shape)
```

```python
Y = X_standardized @ W + b
print("original Y: ",Y.shape)
```

```python
Y_mean = Y.mean(axis=0, keepdims=True) 
Y_std = Y.std(axis=0, keepdims=True) 
Y_norm = (Y - Y_mean) / (Y_std + 1e-8)
print("Y_norm: ",Y_norm.shape)
```

```python
X_new = X @ W 
print("X_raw:", X.shape, "→ X_new:", X_new.shape)
```
